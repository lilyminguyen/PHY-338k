### C. comparison with theory

a. circuit A

```python
# theoretical | V_out / V_in | 

f = np.logspace(1, 6, 500)
omega = 2 * np.pi * f

R = 20e3
C=0.005e-6

gain = 1 / np.sqrt(1 + (omega * R * C)**2)

plt.figure(figsize=(6,4))
plt.semilogx(f, gain, color='#A2C2E0')
plt.axhline(0.707, color='0.7', linestyle='--', label='3 dB cutoff (Gain = 0.707)')
plt.title('Theoretical Gain vs. Frequency (Low-Pass Filter)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')
plt.grid(alpha=0.4)
plt.legend()
plt.savefig('2.3.a.gain.png')
plt.show()
```

```python
# theoretical phase vs. f (low pass filter)
# f = np.linspace(1,35000,500)
phi = -np.arctan(omega * R * C) * 180 / np.pi

plt.figure(figsize=(6,4))
plt.semilogx(f, phi, color='#A2C2E0')
plt.title('Theoretical Phase Shift vs. Frequency (Low-Pass Filter)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase Shift (degrees)')
plt.grid(alpha=0.4)
plt.savefig('2.3.a.phase.png')
plt.show()
```

b. Circuit B high pass

```python
# gain = 1 / np.sqrt(1 + (omega * R * C)**2)
gain = (omega * C * R)**2 / (1 + (omega * C * R)**2)

plt.figure(figsize=(6,4))
plt.semilogx(f, gain, color='#A2C2E0')
plt.axhline(0.707, color='0.7', linestyle='--', label='3 dB cutoff (Gain = 0.707)')
plt.title('Theoretical Gain vs. Frequency (High-Pass Filter)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')
plt.grid(alpha=0.4)
plt.legend()
plt.savefig('2.3.b.gain.png')
plt.show()
```

```python
# high pass theory phase shift vs. f

phi = 90 - (np.arctan(omega * R * C) * 180 / np.pi)

plt.figure(figsize=(6,4))
plt.semilogx(f, phi, color='#A2C2E0')
plt.title('Theoretical Phase Shift vs. Frequency (High-Pass Filter)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase Shift (degrees)')
plt.grid(alpha=0.4)
plt.savefig('2.3.b.phase.png')
plt.show()
```
