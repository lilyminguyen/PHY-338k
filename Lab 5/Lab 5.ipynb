# Lab 5: Bipolar Junction Transistors

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 2. Basic transistor circuit

a. $I_C$ vs. $V_{CE}$ at fixed $V_{in}$

```python
# V_BE vs. I_C
I_C = np.array([0.0301, 6.75, 14.41, 22.08, 29.71, 37.67, 52.5, 67.9, 71.1])
V_CE = np.array([-0.0121, -0.0818, -0.114, -0.143, -0.172, -0.209, -0.274, -1.067, -1.35])
V_CE = [-v for v in V_CE]  # flip because we reversed the leads in lab
I_B = 0.444 #mA

plt.figure(figsize=(6, 4))
plt.plot(V_CE, I_C, color='#A2C2E0', label='$I_C$ vs $V_{CE}$')
plt.xlabel('$V_{CE}$ (V)')
plt.ylabel('$I_C$ (mA)')
plt.title('Collector Current vs Collector-Emitter Voltage')
plt.grid(alpha=0.4)
plt.tight_layout()
plt.savefig('5.2a.png')
plt.show()
```

```python
# h_FE vs. V_CE
# h_FE = I_C/I_B
h_FE = I_C / I_B

plt.figure(figsize=(6, 4))
plt.plot(V_CE, h_FE, color='#A2C2E0', label='$h_{FE}$ vs $V_{CE}$')
plt.xlabel('$V_{CE}$ (V)')
plt.ylabel('$h_{FE}$')
plt.title('Current Gain vs. Collector-Emitter Voltage')
plt.grid(alpha=0.4)
plt.tight_layout()
plt.savefig('5.2b.png')
plt.show()
```

b. $I_C$ vs. $V_{BE}$ at fixed $V_{CC}$

```python
# V_BE vs. I_C (V_BE on linear scale, I_C on log scale)
I_C = np.array([0.1, 1.0, 5.0, 15.0, 20.3, 3.4, 42.7, 56.7, 89.6, 92.6, 93.8])
V_BE = np.array([0.549, 0.614, 0.655, 0.671, 0.668, 0.661, 0.666, 0.675, 0.731, 0.747, 0.763])

sorted_idx = np.argsort(I_C)
I_C = I_C[sorted_idx]
V_BE = V_BE[sorted_idx]

plt.figure(figsize=(6, 4))
plt.semilogx(I_C, V_BE, linestyle='-', color='#A2C2E0')
plt.xlabel(r'$I_C$ (mA)')
plt.ylabel(r'$V_{BE}$ (V)')
plt.title(r'Base-Emitter Voltage vs. Collector Current')
plt.grid(alpha=0.4)
plt.tight_layout()
plt.savefig('5.2c.png')
plt.show()
```
