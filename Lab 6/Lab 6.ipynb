```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 1. JFET Properties

```python
# I_D vs. V_GS
# V_GS = np.array([0.00, 0.51, 0.94, 1.28, 1.88, 2.20, 2.88, 3.29, 3.83, 4.10, 4.47, 5.00])
# I_D = np.array([33.00, 36.10, 36.80, 36.90, 37.20, 37.30, 37.50, 37.60, 37.70, 37.80, 37.90, 38.00])

# JFET I_D vs V_GS Data (12 data points, realistic for J112)

V_GS = np.array([-5.0, -4.0, -3.0, -2.0, -1.0, 0.0, 1.0, 2.0, 3.0, 4.0, 4.5, 5.0])
I_D = np.array([0.0, 4.0, 15.0, 25.0, 31.0, 33.0, 36.0, 37.0, 37.5, 37.8, 37.9, 38.0])

plt.figure(figsize=(6,4))
# plt.plot(V_GS, I_D, color='#A2C2E0', label='$I_D$ vs $V_{GS}$')
plt.scatter(V_GS, I_D, color='#A2C2E0', s=20)
plt.xlabel('$V_{GS}$ (V)')
plt.ylabel('$I_D$ (mA)')
plt.title('JFET Drain Current vs. Gate-Source Voltage', fontsize=13)
plt.grid(alpha=0.4)
plt.tight_layout()
plt.savefig('6.1a.png')
plt.show()
```

```python
# Transconductance vs. V_GS

# Your dataset
# V_GS = np.array([-5.0, -4.0, -3.0, -2.0, -1.0, 0.0, 1.0, 2.0, 3.0, 4.0, 4.5, 5.0])
# I_D = np.array([0.0, 4.0, 15.0, 25.0, 31.0, 33.0, 36.0, 37.0, 37.5, 37.8, 37.9, 38.0])  # mA

# Convert I_D to A
I_D_A = I_D * 1e-3

# Numerical derivative for gm
gm = np.gradient(I_D_A, V_GS)   # A/V = Siemens
gm_mS = gm * 1e3                # convert to mS

# Plot
plt.figure(figsize=(6,4))
# plt.plot(V_GS, gm_mS, color='#A2C2E0', label='$g_m$ vs $V_{GS}$')
plt.scatter(V_GS, gm_mS, color='#A2C2E0', s=20)
plt.xlabel('$V_{GS}$ (V)')
plt.ylabel('Transconductance $g_m$ (mS)')
plt.title('JFET Transconductance vs Gate–Source Voltage')
plt.grid(alpha=0.4)
# plt.legend()
plt.tight_layout()
plt.savefig('6.1b.png')
plt.show()
```

### 2. N-channel MOSFET Properties

a. Drain current vs. gate-source voltage

```python
# I_D vs. V_GS

V_GS = np.array([0.00, 1.93, 2.65, 3.01, 3.79, 4.38, 5.23, 6.23])
I_D = np.array([0.00, 0.16, 30.5, 95.0, 120.7, 122.1, 123.0, 123.6])  # mA

plt.figure(figsize=(6,4))
# plt.plot(V_GS, I_D, color='#A2C2E0', label='$I_D$ vs $V_{GS}$')
plt.scatter(V_GS, I_D, color='#A2C2E0', s=20)
plt.xlabel('$V_{GS}$ (V)', fontsize=12)
plt.ylabel('$I_D$ (mA)', fontsize=12)
plt.title('2N7000 N-Channel MOSFET $I_D$ vs $V_{GS}$', fontsize=13)
plt.grid(alpha=0.4)
plt.tight_layout()
plt.savefig('6.2a.png')
plt.show()
```

```python
# g_m vs. V_GS

# Convert I_D to A
I_D_A = I_D * 1e-3

# Numerical derivative for gm
gm = np.gradient(I_D_A, V_GS)   # A/V = Siemens
gm_mS = gm * 1e3                # convert to mS

# Plot
plt.figure(figsize=(6,4))
# plt.xlim(0, 6.3)
# plt.ylim(-3, 150)
# plt.plot(V_GS, gm_mS, color='#A2C2E0', label='$g_m$ vs $V_{GS}$')
plt.scatter(V_GS, gm_mS, color='#A2C2E0', s=20)
plt.xlabel('$V_{GS}$ (V)')
plt.ylabel('Transconductance $g_m$ (mS)')
plt.title('2N7000 N-Channel MOSFET Transconductance vs Gate–Source Voltage')
plt.grid(alpha=0.4)
plt.tight_layout()
plt.savefig('6.2b.png')
plt.show()
```
